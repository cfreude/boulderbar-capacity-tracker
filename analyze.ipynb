```python
import pandas as pd
import datetime
```

```python
path = "./boulderbar-capacity-log.csv"
df = pd.read_csv(
    path,
    header=0,
    sep=',',
    na_values = {"-1",""})
df['Date']= pd.to_datetime(df['Date'], format="%d-%m-%Y %H:%M:%S", yearfirst=False, dayfirst=True)
df = df.set_index('Date')
```

```python
print(df.info())
print(df)
```

```python
df = df.drop(columns=['Linz', 'Salzburg'])
print(df.info())
```

```python
import plotly.express as px
fig = px.line(df, x=df.index, y=df.columns)
fig.show()
```

```python
# mask the hours we want
hours = df.index.hour
mask = (hours >= 9) & (hours <= 23)
odh = df[mask]

daily_mean = odh[odh.columns].resample('D').mean()
#print(daily_mean)
fig = px.line(daily_mean)
fig.show()
```
